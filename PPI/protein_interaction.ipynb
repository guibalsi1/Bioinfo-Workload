```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Carregar dados
df = pd.read_csv('9606.protein.physical.links.detailed.v12.0.txt', sep=' ')
```

```python
# Filtrar interações humanas
df_human = df[df['protein1'].str.startswith('9606.') & df['protein2'].str.startswith('9606.')].copy()

# Remover o prefixo '9606.' para simplificar
df_human['protein1'] = df_human['protein1'].str.replace('9606.', '')
df_human['protein2'] = df_human['protein2'].str.replace('9606.', '')
```

```python
# Criar grafo
G = nx.from_pandas_edgelist(df_human, 'protein1', 'protein2')
```

```python
# Desenhar o grafo
plt.figure(figsize=(12, 10))
nx.draw(G, with_labels=False, node_size=10)
plt.title('Rede de Interações Proteína-Proteína (Homo sapiens)')
plt.show()
```
